```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

order_items_path='/content/drive/MyDrive/Jet2premium/order_items.csv'
promo_path='/content/drive/MyDrive/Jet2premium/promotions.csv'
df_items=pd.read_csv(order_items_path)
df_promo=pd.read_csv(promo_path)

print("="*30)
print("PHẦN 1: DATA AUDIT")
print("="*30)
# Kiểm tra cấu trúc và kiểu dữ liệu (Type)
print("\n1. Thông tin cấu trúc & Kiểu dữ liệu:")
df_items.info()
#Kiểm tra missing value
print("\n2. Số lượng giá trị thiếu mỗi cột:")
print(df_items.isnull().sum())
#duplicates and cardinality check
print("Số dòng trùng lặp:", df_items.duplicated().sum())
print("\n Số lượng giá trị duy nhất - cardinality:")
print(df_items.nunique())

print("\n" + "="*30)
print("PHẦN 2: GIẢI QUYẾT CÂU 5")
print("="*30)

# Tổng số dòng trong order_items
tong = len(df_items)
print(f"Tổng số dòng: {tong:,}")
# Lọc các dòng có promo_id KHÔNG BỊ NULL (tức là có áp dụng khuyến mãi)
promo_applied = df_items['promo_id'].notna().sum()
print(f"Số sản phẩm có áp dụng khuyến mãi: {promo_applied:,}")
# Tính tỷ lệ phần trăm
percentage = (promo_applied / tong) * 100
print(f"==> Tỷ lệ % các sản phẩm có áp dụng khuyến mãi là: {percentage:.2f}%, ĐÁP ÁN B.39%")
```
