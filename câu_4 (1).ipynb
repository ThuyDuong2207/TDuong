```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.float_format', '{:.2f}'.format)

drive_file_path = '/content/drive/MyDrive/Jet2premium/web_traffic.csv'

df = pd.read_csv(drive_file_path)
print(f"Row: {len(df):,} / Cols: {df.shape[1]}")
print(df.head())

print("="*30)
print("PHẦN 1: DATA AUDIT")
print("="*30)
#chuyển data sang datatime
df['date'] = pd.to_datetime(df['date'])
# Kiểm tra cấu trúc và kiểu dữ liệu (Type)
print("\n1. Thông tin cấu trúc & Kiểu dữ liệu:")
df.info()
#Kiểm tra missing value
print("\n2. Số lượng giá trị thiếu mỗi cột:")
print(df.isnull().sum())

#duplicates and cardinality check
print("Số dòng trùng lặp:", df.duplicated().sum())
print(df.nunique())

# Số loại traffic source
print(f"\ntraffic_source ({df['traffic_source'].nunique()} loại):")
for src in df['traffic_source'].unique():
    print(f"   • {src}")
print(f"\nSố ngày không bị trùng trong data : {df['date'].nunique():,}")

# Kiểm tra có giá trị âm hay không
num_cols = ['sessions', 'unique_visitors', 'page_views',
            'bounce_rate', 'avg_session_duration_sec']
print(df[num_cols].describe().round(4).to_string())
print("\nKiểm tra giá trị âm:")
for col in num_cols:
    neg = (df[col] < 0).sum()
    status = f"{neg} giá trị âm" if neg else "Hợp lệ"
    print(f"   {col:<30}: {status}")

# Kiểm tra Leakage & Time Imbalance
print("\n LEAKAGE & TIME IMBALANCE:")

print("\n" + "="*30)
print("PHẦN 2: GIẢI QUYẾT CÂU 4")
print("="*30)
bounce_rate= df.groupby('traffic_source')['bounce_rate'].mean().sort_values()

print("\nTỷ lệ bounce_rate trung bình theo từng web_traffic:")
print(bounce_rate)

#tỷ lệ thấp nhất
min_source = bounce_rate.idxmin()
min_value = bounce_rate.min()

print(f"\n==> Nguồn '{min_source}' có tỷ lệ thoát trung bình thấp nhất ({min_value:.6f})")
```
